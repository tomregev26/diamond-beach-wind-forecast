```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

date_column_label = "Date"
df = pd.read_csv('../data/processed/final.csv')
df[date_column_label] = pd.to_datetime(df[date_column_label])
df.head()
```

```python
# Data Split Step

import pandas as pd

year_column_label = 'year'
df[year_column_label] = df[date_column_label].dt.year

train_period = (df[year_column_label] >= 2009) & (df[year_column_label] <= 2018)
val_period = (df[year_column_label] >= 2019) & (df[year_column_label] <= 2020)
test_period = (df[year_column_label] >= 2021) & (df[year_column_label] <= 2022)

# Split 
train_df = df[train_period].drop(columns=[year_column_label])

#y_train = df[train_period]['Good Surfing']  
val_df = df[val_period].drop(columns=[year_column_label])

#y_val = df[val_period]['Good Surfing']  
test_df = df[test_period].drop(columns=[year_column_label]) 
#y_test = df[test_period]['Good Surfing']  

# Print the shapes of the datasets
print(f"train shape: {train_df.shape}")
print(f"val shape: {val_df.shape}")
print(f"test shape: {test_df.shape}")
```

```python
# save split
train_df.to_csv('../data/processed/final_train.csv', index=False)
val_df.to_csv('../data/processed/final_val.csv', index=False)
test_df.to_csv('../data/processed/final_test.csv', index=False)
```
