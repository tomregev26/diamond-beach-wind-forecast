```python
import pandas as pd
date_column_label = "Date"
df = pd.read_csv("../data/processed/none.csv")
df[date_column_label] = pd.to_datetime(df[date_column_label])
df.head()
```

```python
df.info()
```

```python
print("Shape", df.shape)
print("Columns" , df.columns)
```

```python
# Degree Normilization
import numpy as np

deg_df = df.copy()
direction = 'wind direction'

dir_columns = [col for col in deg_df.columns if direction in col.lower()]

for col in dir_columns:
    deg_df[f'sin({col})'] = np.sin(np.deg2rad(deg_df[col]))
    deg_df[f'cos({col})'] = np.cos(np.deg2rad(deg_df[col]))

deg_df.drop(columns=dir_columns, inplace=True)
deg_df.info()
```

```python
final_df = deg_df
print(final_df.columns)
final_df.to_csv('../data/processed/final.csv', index=False)
```
